```python
# Finding steady states and threshold current in FHN.
# This cell will show the roots for the FHN model and 7 x 3 current plots in a large figure.

FHN_neuron = FHN_Neuron() # default params

currents = np.linspace(0,2,num = 21) # array of currents to pass to FHN_neuron

# For loop to create a large figure overview of FHN plots
fig = plt.figure(figsize=(15, 20))
fig.subplots_adjust(hspace=0.6, wspace=0.2)
for i , current in enumerate(currents):
    
    ax = fig.add_subplot(7, 3, i+1)
    FHN_neuron.plot( FHN_neuron.solve(round(current,2)) , ax=ax, save=False)


# fig.savefig("Figures/FHN_Range.eps", format='eps')

FHN_neuron.solve_ss() # solve roots and print
```

```python
# This cell is used to investigate each parameter.
# Plots two plots, one with defined params and one with default params.

# Parameters only for FHN_Neuron_1
a = 0.7
b = 0.8
tau = 12.5
I  = 0.325

FHN_neuron_1 = FHN_Neuron(a = a, b = b, tau = tau)
FHN_neuron_2 = FHN_Neuron() # Basic FHN neuron with default parameters


plot_1 = FHN_neuron_1.plot(FHN_neuron_1.solve(I), save=[False,"nameoffile"])
plot_2 = FHN_neuron_2.plot(FHN_neuron_2.solve(I), save=[False, "defaultparamFHN"])

fig, axs = plt.subplots(2)
FHN_neuron_1.plot_nullclines(I, ax=axs[0])

FHN_neuron_2.plot_nullclines(I, ax=axs[1])
# plt.show()

# fig, ax = plt.subplots()
# plot_nullclines(I=0, ax=ax)


# save=[False,'nullcline_threshold']
# title = ("Nullclines: I = %s, a = %s, b = %s, tau = %s" % (I, a, b, tau))
# print(title)
# plt.title(title)
# plt.savefig('Figures/'+save[1]+'.eps', format='eps')
```

# Rinzel Neuron

```python
# Finding steady states and threshold current in Rinzel model.
# This cell will show the roots for the Rinzel model and returns individual Rinzel model plots with range of currents.
# Running this cell will take some time.

R_neuron = Rinzel_Neuron()

currents = np.linspace(0,2,num = 21)

for i , current in enumerate(currents):
    R_neuron.plot( R_neuron.solve(round(current,2)), save=[False,'R_Range/R_Range_'+str(current)])


R_neuron.solve_ss()
```

```python
# This cell plots the Rinzel model at threshold current with default params.
I = 0.33
R_neuron.plot( R_neuron.solve(I) , save=[False,'defaultparamsR'])
```

```python
# This cell is used to investigate each parameter.
# Plots two plots of two subplots (for v,w and z), one with defined params and one with default params.

# Parameters for R_neuron_1 only
a = 0.7
b = 0.8
tau = 12.5

e = 0.0001
c = -0.775

I  = 0.33 # Threshold current


R_neuron_1 = Rinzel_Neuron(a=a, b=b, tau=tau, e=e, c=c)
R_neuron_1.plot(R_neuron_1.solve(I),save=[False,"Rinzel/c/R_c_%s" % (c)])


R_neuron_2 = Rinzel_Neuron() # Basic Rinzel neuron with deafult params
R_neuron_2.plot(R_neuron_2.solve(I), save=[False,"Rinzel/a/R_a_%s" % (a)])

```
